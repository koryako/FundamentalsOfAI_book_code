```python
# -*- coding:utf-8 -*-
import pandas as pd
import random
import fasttext
import jieba
from sklearn.model_selection import train_test_split
```

```python
def getStopWords(datapath):
    stopwords=pd.read_csv(datapath,index_col=False,quoting=3,sep="\t",names=['stopword'], encoding='utf-8')
    stopwords=stopwords["stopword"].values
    return stopwords
```

```python
stopwords=getStopWords('stopword.txt')
```

```python
### #利用pandas把数据读进来
df_technology = pd.read_table("toutiao_cat_data.txt",sep="_!_",encoding ="utf-8",header=None)
#train_x,train_y=df_technology.pop(3),df_technology.pop(2)
print(train_x.shape)
#print(train_x)
catelog_dic={'news_story':100,
            'news_culture':101,
             'news_entertainment':103,
             'news_sports':104,
             'news_finance':105,
             'news_house':106,
             'news_car':107,
             'news_edu':108,
             'news_tech':109,
             'news_travel':110,
             'news_world':111,
             'news_agriculture':112,
             'news_game':113,
            }


sentences=[]
#for index, row in df.iterrows():
    #print row["c1"], row["c2"]
for index,row in df_technology.iterrows():
    line,label=row[3],row[2]
    #print(line)
    try:
        segs=jieba.lcut(line)    #利用结巴分词进行中文分词
        segs=filter(lambda x:len(x)>1,segs)    #去掉长度小于1的词
        #segs=filter(lambda x:x not in stopwords,segs)    #去掉停用词
        sentences.append("__label__"+str(catelog_dic[label])+" , "+" ".join(segs))    #把当前的文本和对应的类别拼接起来，组合成fasttext的文本格式
    except Exception as e:
        print (line)
        continue
    
```

```python
print (len(sentences))
```

```python
def writeData(sentences,fileName):
    print("writing data to fasttext format...")
    out=open(fileName,'w')
    for sentence in sentences:
        out.write(sentence.encode('utf8')+"\n")
    print("done!")
```

```python
writeData(sentences,'train.txt')
```
